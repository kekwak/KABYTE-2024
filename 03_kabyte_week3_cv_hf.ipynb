````python
dataset = load_dataset('fashion_mnist')
dataset['train'][100]
````

````python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
device
````

````python
dataset['train'] = dataset['train'].shuffle(seed=42).select(range(2000))
dataset['test'] = dataset['test'].shuffle(seed=42).select(range(200))
````

````python
class_names = dataset['train'].features['label'].names
id2label = {idx: name for idx, name in enumerate(class_names)}
label2id = {name: idx for idx, name in enumerate(class_names)}
id2label, label2id
````

````python
processor = AutoImageProcessor.from_pretrained('microsoft/resnet-50')

def processor_function(examples):
    return processor(images=list(map(load_image, examples['image']))).data

dataset = dataset.map(processor_function, batched=True)

small_train_dataset = dataset['train']
small_eval_dataset = dataset['test']
````

````python
model = AutoModelForImageClassification.from_pretrained(
    'microsoft/resnet-50',  # './test_trainer/checkpoint-5000'  # load pretrained
    num_labels=10,
    ignore_mismatched_sizes=True,
    id2label=id2label,
    label2id=label2id,
).to(device)
````

````python
training_args = TrainingArguments(output_dir='test_trainer')
metric = evaluate.load('accuracy')

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)

training_args = TrainingArguments(
    num_train_epochs=20,
    output_dir='test_trainer',
    evaluation_strategy='epoch',
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=small_train_dataset,
    eval_dataset=small_eval_dataset,
    compute_metrics=compute_metrics,
)
````

````python
trainer.train()
````

````python
ix = np.random.randint(0, len(dataset['test']))
image, label = dataset['test'][ix]['image'], dataset['test'][ix]['label']

plt.imshow(image, cmap='gray')
plt.show()

inputs = processor(images=load_image(image), return_tensors='pt').to(device)

with torch.no_grad():
    outputs = model(**inputs)
    logits = outputs.logits

predictions = torch.argmax(logits, dim=-1)
print(f'Predicted class: {model.config.id2label[predictions.item()]}\nTrue class: {model.config.id2label[label]}')
````

2\. Создайте новый [`pipeline`](https://huggingface.co/docs/transformers/v4.39.0/en/main_classes/pipelines#pipeline-custom-code) на основе любой из обученных в предыдущем задании моделей. Продемонстируйте пример работы на нескольких изображениях.

````python
class CustomImageClassificationPipeline(ImageClassificationPipeline):
    def __init__(self, model, processor):
        super().__init__(model)

        self.model = model
        self.processor = processor

        self.id2label = self.model.config.id2label
        self.label2id = self.model.config.label2id

    def predict(self, images):
        inputs = self.processor(
            images=list(map(load_image, images)),
            return_tensors='pt'
        )

        with torch.no_grad():
            outputs = self.model(**inputs)
            logits = outputs.logits

        predictions = torch.argmax(logits, dim=-1)
        
        return predictions
    
    def __call__(self, *args, **kwargs):
        return self.predict(*args, **kwargs)

my_pipeline = CustomImageClassificationPipeline(model=model, processor=processor)
````

````python
ix = np.random.choice(len(dataset['test']), 4)
image, label = dataset['test'][ix]['image'], dataset['test'][ix]['label']

fig, ax = plt.subplots(nrows=1, ncols=len(ix), figsize=(8, 8))
for i, j in enumerate(image):
    ax[i].imshow(j, cmap='gray')

predictions = my_pipeline(image)

print(f'True classes:      {label}')
print(f'Predicted classes: {predictions.tolist()}')

plt.show()

print(f"№  {'True':^20} | {'Predicted':^20} | {'Is True':^8}")
for i, (j, k) in enumerate(zip(label, predictions)):
    print(f'{i}. {my_pipeline.id2label[j]:^20} | {my_pipeline.id2label[k.item()]:^20} | {j == k.item():^8}')
````
